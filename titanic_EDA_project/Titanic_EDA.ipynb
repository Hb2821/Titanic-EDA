```python
df["Age"] = df.groupby(["Sex","Pclass"])["Age"].transform(
    lambda x: x.fillna(x.median())
)
print(df["Age"].isnull().sum())
```

**Survival chance by gender**

```python
survival_chance = df.groupby("Sex")["Survived"].mean()*100

print(survival_chance)
```

```python
plt.figure(figsize=(6,4))

plt.bar(
    survival_chance.index,
    survival_chance.values
)

plt.xlabel("Gender")
plt.ylabel("Survival Chance (%)")
plt.title("Titanic Survival Chance by Gender")

plt.ylim(0,100)

for i,value in enumerate(survival_chance):
    plt.text(i,value+2,f"{value:.2f}%")

plt.show()
```

**Survival chance by Class**

```python
class_survival = df.groupby("Pclass")["Survived"].mean() * 100

print(class_survival)
```

```python
plt.figure(figsize=(6,4))

plt.bar(
    class_survival.index.astype(str),
    class_survival.values
)

plt.xlabel("Passenger Class")
plt.ylabel("Survival Chance (%)")
plt.title("Titanic Survival Chance by Class")

plt.ylim(0,100)
```

**Survival analysis by age and gender.**

```python
df["Age_Group"] = pd.cut(
    df["Age"],
    bins=[0,12,30,50,100],
    labels=["Child","Young","Adult","Senior"]
)
```

```python
age_sex_survival = df.groupby(
    ["Age_Group","Sex"],
    observed=True
)["Survived"].mean()*100


print(age_sex_survival)
```

```python
plt.figure(figsize=(8,5))


age_sex_survival.unstack().plot(
    kind="bar"
)


plt.xlabel("Age Group")
plt.ylabel("Survival Chance (%)")
plt.title("Survival Chance by Age Group and Gender")

plt.ylim(0,100)

plt.legend(title="Gender")

plt.show()
```

**Titanic EDA Final Observations:**

1. Female passengers had a higher survival rate compared to male passengers, showing the priority given to women during rescue.

2. Male passengers had a lower survival chance, as many men were left behind while saving women and children.

3. First-class passengers had better survival chances compared to second and third-class passengers due to better access to lifeboats.

4. Younger passengers, especially children, had higher survival chances compared to many adult passengers.

5. Survival was mainly affected by **gender, age, and passenger class**.
